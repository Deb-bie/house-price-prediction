```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
# Load Data
df = pd.read_csv('../data/raw/dataset_house.csv')
```

```python
df.dtypes.value_counts()
```

```python
numerical = df.select_dtypes(include=[np.number]).columns
print(f"\nNumerical Features: {len(numerical)}")
```

```python
categorical = df.select_dtypes(include=['object']).columns
print(f"\nCategorical Features: {len(categorical)}")
```

```python
# Missing Values
missing = df.isnull().sum()
print(missing)
```

```python
# Correlation Heatmap
plt.figure(figsize=(14, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.savefig("../reports/correlation_matrix.png")
plt.show()
```

```python
top_corr = corr['price'].sort_values(ascending=False).head(11)[1:]
print("\nTop Features Correlated with Price:")
print(top_corr)
```

```python
stats.probplot(df['price'], plot=plt)
plt.title('Price QQ-Plot')
plt.savefig('../reports/target_distribution.png')
plt.show()
```

```python
# Distribution of Target Variable
sns.histplot(df['price'], kde=True)
plt.title("Price Distribution")
plt.savefig('../reports/target_distribution_kde.png')
plt.show()
```

```python
# transforming the price using log to compress the high values and spread out the lower ones.
df['log_price'] = np.log1p(df['price'])  # log(1 + price)
```

```python
stats.probplot(df['log_price'], plot=plt)
plt.title('Log-Transformed Price QQ-Plot')
plt.savefig('../reports/log_transformed_target_distribution.png')
plt.show()
```

```python
sns.histplot(df['log_price'], kde=True)
plt.title("Log-Transformed Price Distribution")
plt.savefig('../reports/log_transformed_kde_target_distribution.png')
plt.show()
```

```python
df.to_csv('../data/processed/cleaned_data.csv', index=False)
print("\nEDA completed. Cleaned data saved!")
```
